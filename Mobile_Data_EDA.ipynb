```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./train.csv')
```

```python
data
```

```python
data.isnull().sum()
```

### Description of features:

#### 1- battery_power: Total energy a battery can store in one time measured in (mAh)

#### 2- blue: Has bluetooth or not

#### 3- clock_speed: Speed at which microprocessor executes instructions

#### 4- dual_sim: Has dual sim support or not

#### 5- fc: Front camera (Megapixels)

#### 6- four_g: Has 4G or not

#### 7- int_memory: Internal memory in (Gigabytes)

#### 8- m_dep: Mobile depth in (Cm)

#### 9- mobile_wt: Weight of mobile phone

#### 10- pc: Primary camera (Megapixels)

#### 11- px_height: Pixel resolution height

#### 12- px_width: Pixel resolution width

#### 13- ram: Random access memory in (Megabytes)

#### 14- sc_h: Screen height of mobile in (Cm)

#### 15- sc_w: Screen width of mobile in (Cm)

#### 16- talk_time: Longest time that a single battery charge will last when you are constantly talking on the phone

#### 17- three_g: Has 3G or not

#### 18- touch_screen: Has touch screen or not

#### 19- wifi: Has wifi or not

#### 20- n_cores: Number of cores of processor

#### 21- price_range: This is the Target variable with value of 0: (Low Cost), 1: (Medium Cost), 2: (High Cost), and 3: (Very High Cost)

n_cores and CLOCK Speed

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=data, x='n_cores', y='clock_speed', color='teal')

plt.title('Distribution of clock speed by n_cores')
plt.show()
```

##### They all have almost similar Clock speed irrigardless of the n_cores, meaning they have no eefect on each other hence no relation

```python
plt.figure(figsize=(10,8))

sns.scatterplot(data=data, x='n_cores', y='ram')
plt.title('Scatter of ram against n_cores')
plt.plot()
```

##### No clear relation between n_cores and ram since there is the same distribution in all of the n_cores values

```python
plt.figure(figsize=(10,10))

sns.scatterplot(data=data, x='touch_screen' , y='ram', hue='price_range')
plt.title('Relation between ram and a touch screen')
```

##### No connection between the touch screen and ram
##### Meaning a touch screen device does not require more ram to handle the touch screen features

```python
fig = plt.figure(figsize=(10,10))
scatter = plt.scatter(data=data, x='px_height', y='px_width', c='price_range', cmap='coolwarm')

legend1 = plt.legend(*scatter.legend_elements(), title="Price Range")
plt.gca().add_artist(legend1)
plt.title('Relation between px_width and px_height')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr())
plt.title('Correaltion Heatmap of the data')

plt.show()
```

#### Price distribution

```python
plt.figure(figsize=(10,10))
grouped = data.groupby('price_range').count()['blue']

plt.pie(grouped.values, labels=grouped.index, autopct='%1.1f%%', wedgeprops = { 'linewidth' : 4, 'edgecolor' : 'white' })
plt.show()
```

```python
data.hist(bins=30, figsize=(20, 20))
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.title('Boxplots of Numerical Features')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
scatter = plt.scatter(data['ram'], data['battery_power'], c=data['price_range'], cmap='coolwarm')
plt.xlabel('RAM')
plt.ylabel('Battery Power')
plt.title('RAM vs. Battery Power vs. Price Range')
plt.legend()

legend1 = plt.legend(*scatter.legend_elements(), title="Price Range")
plt.gca().add_artist(legend1)

plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.kdeplot(data=data, x='ram', label='ram')
sns.kdeplot(data=data, x='battery_power', label='battery power')
sns.kdeplot(data=data, x='px_height', label='px_height')
sns.kdeplot(data=data, x='px_width', label='px_width')
plt.title('Distribution of the strongest related features')

plt.legend()

plt.show()
```
